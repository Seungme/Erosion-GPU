```python
import csv
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
gpu = pd.read_csv('GPU.csv', sep=',',header=None)
gpu = gpu.values
cpu = pd.read_csv('CPU.csv', sep=',',header=None)
cpu = cpu.values
cpuv = pd.read_csv('CPUVectorized.csv', sep=',',header=None)
cpuv = cpuv.values

cpuKer = pd.read_csv('CPUKerSize.csv', sep=',',header=None)
cpuKer = cpuKer.values
cpuvKer = pd.read_csv('CPUVectorizedKerSize.csv', sep=',',header=None)
cpuvKer = cpuvKer.values
gpuKer = pd.read_csv('GPUKersize.csv', sep=',',header=None)
gpuKer = gpuKer.values

plt.style.use('fivethirtyeight')
```

```python
fig, ax = plt.subplots(figsize=(28, 10))
cput = cpu.T
cpuvt = cpuv.T
gput = gpu.T
ax.plot(cput[0], cput[1], label='CPU')
ax.plot(cpuvt[0], cpuvt[1], label='CPU Vectorized')
ax.plot(gput[0], gput[1], label='GPU')

plt.ylabel('time (s)')
plt.xlabel('Image size')
ax.set_title("[Benchmark] Varying image size.")

plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(28, 10))
ax.plot(cpuvt[0], cpuvt[1], label='CPU Vectorized')
ax.plot(gput[0], gput[1], label='GPU')
plt.ylabel('time (s)')
plt.xlabel('Image size')
ax.set_title("[Benchmark] Varying image size.")

plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(28, 10))
cpuKert = cpuKer.T
cpuvKert = cpuvKer.T
gpuKert = gpuKer.T
ax.plot(cpuKert[0][:15], cpuKert[1][:15], label='CPU')
ax.plot(cpuvKert[0][:15], cpuvKert[1][:15], label='CPU Vectorized')
ax.plot(gpuKert[0][:15], gpuKert[1][:15], label='GPU')
plt.ylabel('time (s)')
plt.xlabel('Kernel size')
ax.set_title("[Benchmark] Varying kernel size.")

plt.legend()
plt.show()
```
