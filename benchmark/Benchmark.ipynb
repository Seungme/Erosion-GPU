```python
import matplotlib.pyplot as plt
import csv

x=[]
y=[]

x2=[]
y2=[]

x3=[]
y3=[]

with open('GPU.csv', 'r') as csvfile:
    plots= csv.reader(csvfile, delimiter=',')
    for row in plots:
        if (len(row) == 0):
            continue
        x.append(float(row[0]))
        y.append(float(row[1]))
        
with open('CPU.csv', 'r') as csvfile:
    plots= csv.reader(csvfile, delimiter=',')
    for row in plots:
        if (len(row) == 0):
            continue
        x2.append(float(row[0]))
        y2.append(float(row[1]))
        
with open('CPUVectorized.csv', 'r') as csvfile:
    plots= csv.reader(csvfile, delimiter=',')
    for row in plots:
        if (len(row) == 0):
            continue
        x3.append(float(row[0]))
        y3.append(float(row[1]))

plt.figure(figsize=(20, 20))
plt.plot(x,y, 'r', marker='o', label="GPU")
plt.plot(x2,y2, 'b',marker='o', label="CPU")
plt.plot(x3,y3, 'g',marker='o', label="CPU vectorized")
plt.legend()

plt.title('Data from the CSV File: People and Expenses')

plt.xlabel('Image size')
plt.ylabel('Execution time')

plt.show()
```
